```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# California Housing Prices
california = fetch_california_housing()

# Reading the dataset
df = pd.DataFrame(california.data, columns=california.feature_names)

# Getting the target values (price) and adding it to the DataFrame
df['PRICE'] = california.target
print("DataFrame")
print(df)
print("")

# Examining the DataFrame
print("Summary of the DataFrame")
print(df.info())
print("")

print("Descriptive statistics of the dataframe")
print(df.describe())
print("")

print("Total number of missing values")
print(df.isnull().sum())
print("")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

TEST_SIZE = 0.4

# Features
X = df.iloc[:, :-1].values
# Target
y = df.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE
    )

model = LinearRegression()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print (f"Accuracy = {accuracy}")
```

```python
from sklearn.metrics import mean_squared_error

accuracy_training = model.score(X_train, y_train)
accuracy_test = model.score(X_test, y_test)
accuracy = model.score(X, y)
print('Accuracy:')
print(f'Accuracy_training = {accuracy_training}')
print(f'Accuracy_test = {accuracy_test}')
print(f'Accuracy = {accuracy}')

predictions_training = model.predict(X_train)
predictions_test = model.predict(X_test)
prediction = model.predict(X)

mse_training = mean_squared_error(y_train, predictions_training)
mse_test = mean_squared_error(y_test, predictions_test)
mse = mean_squared_error(y, prediction)

print('Mean Squared Error:')
print(f'mse_training = {mse_training}')
print(f'mse_test = {mse_test}')
print(f'mse = {mse}')
```
