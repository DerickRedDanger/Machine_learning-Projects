```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Breast Cancer Wisconsin (Diagnostic) Dataset
cancer = load_breast_cancer()

# Reading the dataset
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

#can = pd.DataFrame(cancer.target)

# Getting the target values (Cancer) and adding it to the DataFrame
df['Cancer'] = cancer.target
print("DataFrame")
print(df)
print("")

# Examining the DataFrame
print("Summary of the DataFrame")
print(df.info())
print("")

print("Descriptive statistics of the dataframe")
print(df.describe())
print("")

print("Total number of missing values")
print(df.isnull().sum())
print("")
```

##Training and evaluating an Logistic Regression, using Standard Scaller.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing

TEST_SIZE = 0.4

# Features
X = df.iloc[:, :-1].values
# Target
y = df.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=55
    )

model = LogisticRegression()

scaler = preprocessing.StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

model.fit(X_train, y_train)


from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

prediction = model.predict(X_test)

accuracy = accuracy_score(y_test,prediction)
precision = precision_score(y_test,prediction)
recall = recall_score(y_test,prediction)
F1_score = f1_score(y_test,prediction)

print(f"Accuracy with Standard Scaller")
print(f"Accuracy = {accuracy}")
print(f"Precision = {precision}")
print(f"Recall = {recall}")
print(f"F1_score = {F1_score}")
```
